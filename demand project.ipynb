```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("moni3.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("car color")["driver age"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("car color",fontsize=20)
plt.ylabel("driver age",fontsize=20)
plt.title("DEMAND AND SUPPLY ANALYSIS",fontsize=20)
plt.show()
```
